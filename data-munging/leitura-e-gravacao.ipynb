## leitura

```python
# abrindo arquivo apenas leitura
arq = open("/content/drive/MyDrive/codes/DS/kernels/arquivos/files/teste1.txt","r", encoding='utf8')
```

```python
# lendo
print(arq.read())
```

```python
#note que uma vez realizado uma leitura, o cursor se movimentou e não voltou ao início
print(arq.read())
```

```python
# mover cursor (linha,coluna), a função retorna a posição selecionada
print(arq.seek(0,0))
print(arq.read())
```

```python
# número de carácteres
print(arq.tell())
```

```python
arq.close()
```
